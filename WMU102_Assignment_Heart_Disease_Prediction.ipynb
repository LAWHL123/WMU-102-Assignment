We use boxplots to view the ouliers in aech variable of the dataset.

From the boxplots, we observe that for the **continuous** variables **'trestbps'**, **'chol'**, **'thalach'** and **'oldpeak'** contain outliers, hence we must perform outliers treatment to these variables.

We use **Flooring and Capping** for the outliers treatment.
In this technique, we will do the flooring(e.g 25th percentile) for the lower values and capping(e.g for the 75th percentile) for the higher values. These percentile values will be used for the quantile-based flooring and capping.
The data points that are lesser than the lower whisker are replaced with the lower whisker value and the data points that are greater than the upper whisker are replaced with upper whisker value.

```python
features = ['trestbps', 'chol', 'thalach', 'oldpeak']
for each_feature in features:
  Q1 = df[features].quantile(0.25)
  Q3 = df[features].quantile(0.75)
  IQR = Q3 - Q1
  lower_whisker = Q1 - (1.5*IQR)
  upper_whisker = Q3 + (1.5*IQR)
  df[features] = np.where(df[features]>upper_whisker, upper_whisker, 
                 np.where(df[features]<lower_whisker, lower_whisker, df[features]))
```

# **K-Nearest Neighbors (KNN)**


We train the KNN algorithm. Import KNeighborsClassifier from sklearn.neighbors
library. Then, we need to find the optimal value of k as it can impact the model performance.
We find the value of k by using looping. 

```python
# Fitting classifier to the Training set
# Import the KNeighborsClassifier class from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Import metrics model to check the accuracy
from sklearn import metrics

# Try running from k=1 through 25 and record testing accuracy
k_range = range(1, 26)
scores = {}
scores_list = []
for k in k_range:
    KNN = KNeighborsClassifier(n_neighbors=k)
    KNN.fit(X_train, y_train)
    y_pred = KNN.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test, y_pred)
    scores_list.append(metrics.accuracy_score(y_test, y_pred))
```

```python
scores
```

Then, plot the graph of testing accuracy versus value of k in KNN using the matplotlib library. As we can see there is a raise and fall in the accuracy and it is quite typical when examining the model complexity with the accuracy. In general as the value of K increase there appears to be a raise in the accuracy and again it falls.

In general the training accuracy rises as the model complexity increases, for KNN the model complexity is determined by the value of K. Larger K value leads to smoother decision boundary (less complex model). Smaller K leads to more complex model (may lead to overfitting). Testing accuracy penalizes models that are too complex(over fitting) or not complex enough(underfit). 

```python
# Plot the relationship between K and the testing accuracy

plt.plot(k_range, scores_list)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```


From the graph, we can observe that the first peak point is at K=2, so here we choose 2 as my k_value.

Then, train the KNN algorithm with k=2 and make prediction on the test data.

```python
KNN = KNeighborsClassifier(n_neighbors=2)
KNN.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred1 = KNN.predict(X_test)
y_pred1
```

# **Step 8: Evaluate the Model**
Next step is evaluating the model. Accuracy, recall, precision and confusion matrix are the
most commonly used metrics.

* The **accuracy** is a metric defining the rate at which a model has classified the records correctly.
* The **recall** is the percentage of positive instances that were correctly identified. 
* The **precision** is the percentage of instances marked positive that really are positive.
* The **confusion matrix** is a table that summarizes the classifier’s predictions  against the actual known data categories.

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score

print ('Accuracy:', accuracy_score(y_test, y_pred1))
print ('Recall:', recall_score(y_test, y_pred1, average="weighted"))
print ('Precision:', precision_score(y_test, y_pred1, average="weighted"))
confusion = confusion_matrix(y_test, y_pred1)
print('Confusion matrix:')
print(confusion)
```

The K-Nearest Neighbors (KNN) model has 88.52% accuracy.

##**Comparison between models**

**K-Nearest Neighbours (KNN)** is a simple supervised machine learning algorithm that can be used for both classification and regression predictive problems. It is mostly used to solve classification problems. It stores all the available cases and classifies the new data based on a similarity measure. It is a nonparametric algorithm, as it does not make any assumptions on the data being studied. KNN is also a lazy-learner because it does not make any generalization about the data and “memorizes” the training dataset instead. Additionally, KNN is simple to implement. 

**Support Vector Machine (SVM)** is a popular supervised machine learning algorithm that is used for classification and regression problems. SVM algorithm finds the hyperplane that best differentiates the two classes. It finds the plane that maximizes the margin where the margin is the distance between the line and the data points closest to the line. Maximizing the margin can improve the model performance by decreasing the likelihood of misclassification.
On the other hand, the SVM algorithm has some disadvantages. For instance, SVM does not perform well when the dataset has more noise. Besides that, choosing the appropriate kernel function and hyperparameters tuning is not easy. It is also prone to overfitting.

**Decision Tree** uses a tree structure to specify sequences of decisions and consequences. It starts with a single node, which branches into possible outcomes. Each of those outcomes leads to additional nodes, which branch off into other possibilities, hence a decision tree is a treelike algorithm. Decision tree uses multiple algorithms such as ID3 to decide to split a node into two or more sub-nodes. The creation of sub-nodes increases the homogeneity of resultant sub-nodes. However, one of the main disadvantages of decision tree is the instability because a small change in the data can cause a huge change in the structure of the tree. Also, the overfitting problem is quite normal in the decision tree algorithm.

**Logistic Regression** is a supervised learning algorithm used to predict a dependent categorical target variable. It is one of the predominant classification models to predict a binary target variable. It finds the probability of event success and event failure. It studies the relationship from a given set of labelled data. It learns a linear relationship and then introduces a non-linearity in the form of the Sigmoid function. Logistic Regression has one major assumption which is the linearity between the dependent variable and the independent variables. Hence, it is not good for non-linear problems.

From the results, we can observe that all these four models have a quite high accuracy in this classification problem. Particularly, the KNN algorithm outperforms the other algorithms as KNN has the highest accuracy (0.885). Therefore, KNN is the most effective model in this classification problem.

#**Recommendation**
We can improve the performance of the KNN algorithm by several ways. Firstly, reduce the dimensionality of the dataset because KNN algorithm does not work very well with high dimensional data. Technique like Principal Component Analysis (PCA) can be executed when a KNN model is built. Secondly, the choice of distance metric is vital in the performance of KNN model. Different distance metrics such as Euclidean distance, Manhattan distance and Chebyshev distance can be applied to different classification problems to improve the accuracy
of the model. Thirdly, rescaling the data also helps to improve the model performance. Feature scaling is very important in KNN algorithm. If the model accuracy is low, try rescaling the data to improve the accuracy of the model. Last but not least, choosing the optimal value of k is important in KNN machine learning algorithm as it hugely influences the model performance.
Therefore, some techniques such as looping and cross validation are applied to define the appropriate value of k. (Chauhan, 2019) (source: https://www.kdnuggets.com/2019/07/classifying-heart-disease-using-k-nearest-neighbors.html)

For the SVM algorithm, we can improve its performance by improving the data
preprocessing. Make sure that the data is standardized and scaled properly. Also, optimize the hyperparameters because the hyperparameters can hugely impact the model performance. Choosing the right kernel function plays an important role in improving the model accuracy. Therefore, hyperparameter tuning should be performed properly to improve the model performance. Besides, try to reweigh training instances when the dataset is unbalanced.

For the Decision Tree algorithm, we can improve the performance of the model by tackling the overfitting problem. Decision tree is prone to overfitting and hence dealing with overfitting problems is vital in this algorithm. We can prevent overfitting by stopping splitting the tree at some point. Also, we can generate a complete tree first, then get rid of some branches. (source: https://towardsdatascience.com/construct-a-decision-tree-and-how-to-deal-with-overfitting-f907efc1492d)

For the Logistic Regression algorithm, feature scaling or normalization is vital before training the model. For instance, if there are two variables in which their respective value differ significantly, then the feature with a larger value will dominate the others in Logistic Regression. Also, check for class imbalance in the dataset because Logistic Regression havs a class_weight parameter and setting the dataset to become balanced is very important in this algorithm.
